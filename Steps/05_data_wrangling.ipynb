```python
import re

def split_contact(val):
    val = str(val)
    # Find email
    email_match = re.search(r'[\w\.-]+@[\w\.-]+', val)
    email = email_match.group(0) if email_match else None
    # Remove any leading phone number (with dashes, spaces, parentheses) from email
    if email:
        email = re.sub(r'^[\d\-\s\(\)]+', '', email)
    # Find phone (handles dashes, spaces, parentheses)
    phone_match = re.search(r'(\+?\d[\d\s\-\(\)]{7,}\d)', val)
    phone = phone_match.group(0) if phone_match else None
    return pd.Series({'phone': phone, 'email': email})
```

```python

pat_df[['phone', 'email']] = pat_df['contact'].apply(split_contact)

pat_df[['contact', 'phone', 'email']].head()
```

```python
pat_df.drop(columns = 'contact',inplace = True)
```

```python
tre_df=tre_df[tre_df['dosage_range']!="-"]
```

```python
tre_df['dosage_start'] = tre_df['dosage_range'].str.split('-').str.get(0).str.replace('u','')
tre_df['dosage_end'] = tre_df['dosage_range'].str.split('-').str.get(1).str.replace('u','')
```

```python
tre_df.drop(columns="dosage_range",inplace = True)
```

```python
def camera_extractor(text):

    if 'Quad' in text:
        return 4
    elif 'Triple' in text:
        return 3
    elif 'Dual' in text:
        return 2
    else:
        return 1
```

```python
def front_camera_extractor(text):

    if 'Quad' in text:
        return 4
    elif 'Triple' in text:
        return 3
    elif 'Dual' in text:
        return 2
    elif '-1' in text:
        return -1
    else:
        return 1
```

```python
df1['num_rear_cameras'] = df1['camera'].str.split("&").str.get(0).apply(camera_extractor)
```

```python
df1['num_front_cameras'] = df1['camera'].str.split("&").str.get(1).fillna('-1').apply(front_camera_extractor)
```

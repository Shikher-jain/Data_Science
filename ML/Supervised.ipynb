```python
import pandas as pd

from sklearn.linear_model import LinearRegression,LogisticRegression
```

## Regression

```python
model = LinearRegression()
```

```python
df = pd.read_csv('marks.csv')
df.head()
```

```python
X = df[['study_hours']]
y = df['marks']
```

```python
model.fit(X,y)

hours = float(input("Enter study hours: "))
pred = model.predict([[hours]])

print(pred[0])
```

## KNN Kth - Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
df = pd.read_csv('fruits.csv')
df.head()
```

```python
X = df[['weight_g','size_cm']]
y = df['fruit']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_encoded
```

```python
model.fit(X, y_encoded)
weight = float(input("Enter weight (g): "))
size = float(input("Enter size (cm): "))

pred = model.predict([[weight, size]])
print(le.inverse_transform(pred))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('DT.csv')
df.head()
```

```python
model = DecisionTreeClassifier()
```

```python
X = df[['size','color_shade']]
y = df['fruit']
```

```python
model.fit(X,y)

size = float(input("Enter size: "))
color_shade = float(input("Enter color shade: "))

pred = model.predict([[size,color_shade]])
print(pred[0])
```

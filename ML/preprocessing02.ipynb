```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, QuantileTransformer

# Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Quantile Transformation
# quant = QuantileTransformer(output_distribution='normal')
quant = QuantileTransformer()
X_quant = quant.fit_transform(X)

# Figure
plt.figure(figsize=(18,5))

# Original Features
plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Original Features")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Standard Scaled Features
plt.subplot(1, 3, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap='viridis')
plt.title("StandardScaler")
plt.xlabel("Feature 1 (scaled)")
plt.ylabel("Feature 2 (scaled)")

# Quantile Transformed Features
plt.subplot(1, 3, 3)
plt.scatter(X_quant[:, 0], X_quant[:, 1], c=y, cmap='viridis')
plt.title("QuantileTransformer")
plt.xlabel("Feature 1 (quantile)")
plt.ylabel("Feature 2 (quantile)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, QuantileTransformer

# Generate skewed data
x = np.random.exponential(10, 1000) + np.random.normal(0, 1, 1000)

# Manual standardization
x_manual = (x - np.mean(x)) / np.std(x)

# StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x.reshape(-1, 1)).flatten()

# QuantileTransformer
quant = QuantileTransformer(output_distribution='normal')
x_quantile = quant.fit_transform(x.reshape(-1, 1)).flatten()

# Plot all
plt.figure(figsize=(20,5))

# Original
plt.subplot(1, 4, 1)
plt.hist(x, bins=30, color='skyblue', edgecolor='black')
plt.title("Original Data")

# Manual Scaling
plt.subplot(1, 4, 2)
plt.hist(x_manual, bins=30, color='orange', edgecolor='black')
plt.title("Manual Scaling")

# StandardScaler
plt.subplot(1, 4, 3)
plt.hist(x_scaled, bins=30, color='green', edgecolor='black')
plt.title("StandardScaler")

# QuantileTransformer
plt.subplot(1, 4, 4)
plt.hist(x_quantile, bins=30, color='purple', edgecolor='black')
plt.title("QuantileTransformer")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_samples = 1000
sample_size = 30
distribution_range = (0,1)

samples = np.random.uniform(distribution_range[0],distribution_range[1],(num_samples,sample_size))
samples

sample_mean = np.mean(samples,axis=1)
sample_mean

sns.histplot(sample_mean,bins=20,kde=True)
plt.show()
plt.hist(sample_mean,bins=20)
plt.show()
```

```python
num_samples = 1000
sample_size = 50
lambda_param = 2

samples = np.random.exponential(scale=1/lambda_param,size= (num_samples,sample_size))
samples

sample_mean = np.mean(samples,axis=1)
sample_mean

sns.histplot(sample_mean,bins=20,kde=True)
plt.show()
plt.hist(sample_mean,bins=20,edgecolor='black')
plt.show()
```

```python
num_samples = 1000
sample_size = 50

# Parameters for Poisson Distributions
poisson_lambda = 5

# Parameters for Gamma Distributions
gamma_shape, gamma_scale = 2, 1

# Parameters for Binomial Distributions
binomial_n, binomial_p = 10, 0.5
```

```python
# Generate samples
poisson_samples = np.random.poisson(lam=poisson_lambda, size=(num_samples, sample_size))
gamma_samples = np.random.gamma(shape=gamma_shape, scale=gamma_scale, size=(num_samples, sample_size))
binomial_samples = np.random.binomial(n=binomial_n, p=binomial_p, size=(num_samples, sample_size))
```

```python
# Calculate sample means
poisson_mean = np.mean(poisson_samples, axis=1)
gamma_mean = np.mean(gamma_samples, axis=1)
binomial_mean = np.mean(binomial_samples, axis=1)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Plot histograms for Poisson sample means
axs[0].hist(poisson_mean, bins=20, edgecolor='black', density=True)
axs[0].set_title('Sample Means from Poisson Distribution')
axs[0].set_xlabel('Sample Mean')
axs[0].set_ylabel('Density')

# Plot histograms for Gamma sample means
axs[1].hist(gamma_mean, bins=20, edgecolor='black', density=True)
axs[1].set_title('Sample Means from Gamma Distribution')
axs[1].set_xlabel('Sample Mean')
axs[1].set_ylabel('Density')

# Plot histograms for Binomial sample means
axs[2].hist(binomial_mean, bins=20, edgecolor='black', density=True)
axs[2].set_title('Sample Means from Binomial Distribution')
axs[2].set_xlabel('Sample Mean')
axs[2].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
num_samples = 10000
sample_size = 50

gamma_shape, gamma_scale = 2, 1 

theoretical_mean = gamma_shape * gamma_scale
theoretical_variance = gamma_shape * (gamma_scale ** 2)

samples = np.random.gamma(shape=gamma_shape, scale=gamma_scale, size=(num_samples, sample_size))
sample_means = np.mean(samples, axis=1)

empirical_mean = np.mean(sample_means)
empirical_variance = np.var(sample_means)

print(f"Theoretical Mean: {theoretical_mean}")
print(f"Empirical Mean: {empirical_mean}")
print(f"Theoretical Variance: {theoretical_variance/sample_size}")
print(f"Empirical Variance: {empirical_variance}")
```

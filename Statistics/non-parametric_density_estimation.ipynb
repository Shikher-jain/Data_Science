```python
import numpy as np
import pandas as pd
import random
from numpy.random import normal, uniform, choice
import matplotlib.pyplot as plt
```

```python
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = np.hstack((sample1, sample2))
sample1,sample2,sample
```

```python
plt.hist(sample1, bins=50, alpha=0.5, label='Sample 1', density=True)
plt.hist(sample2, bins=50, alpha=0.5, label='Sample 2', density=True)
plt.hist(sample, bins=50, alpha=0.5, label='Sample', density=True)
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KernelDensity
```

```python
model = KernelDensity(kernel='gaussian', bandwidth=3)
sample = sample.reshape(len(sample), 1) # Reshape for sklearn ex. 1000 samples to 1000x1
model.fit(sample)
# sample
```

```python
values = np.linspace(sample.min(), sample.max(), 100)
values = values.reshape(-1, 1)
# values
```

```python
probabilities = model.score_samples(values) # gives the log values of the probability density function
probabilities
probabilities = np.exp(probabilities) # converting log values to actual 
probabilities
```

```python
plt.hist(sample, bins=50, alpha=0.5, label='Sample 1', density=True)
plt.plot(values, probabilities, color='red')
plt.legend()
plt.show()
```

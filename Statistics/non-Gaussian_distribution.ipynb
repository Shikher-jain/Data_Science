```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = sns.load_dataset('iris')
```

```python
df.head()
```

```python
sns.kdeplot(df['sepal_length'])
```

```python
temp = sorted(df['sepal_length'].tolist())
temp
```

```python
sns.kdeplot(temp)
```

```python
y_quant = []

for i in range(1,101):
    y_quant.append(np.percentile(temp,i))
```

```python
sns.kdeplot(y_quant)
```

```python
samples = np.random.normal(loc=0, scale=1, size=1000)
```

```python
sns.kdeplot(samples)
```

```python
x_quant = []

for i in range(1,101):
    x_quant.append(np.percentile(samples,i))
```

```python
sns.kdeplot(x_quant)
```

```python
sns.scatterplot(x=x_quant, y=y_quant)
```

```python
import statsmodels.api as sm
```

```python
fig = sm.qqplot(df['sepal_length'], line ='45',fit=True)
fig = sm.qqplot(df['sepal_length'], line ='r',fit=True)

plt.title("Q-Q plot of Sepal Length")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.show()
```

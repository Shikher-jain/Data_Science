```python
import numpy as np
import pandas as pd
from scipy.stats import chisquare
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Titanic Dataset/train.csv')
```

```python
class_counts = df['Pclass'].value_counts().sort_index()
class_counts
```

```python
total_passengers = len(df)
expected_counts = total_passengers / 3   # Assuming equal distribution across 3 classes
expected = [expected_counts] * 3

print(f" Total Passenger {total_passengers} \n Expected Count {expected_counts} \n Expected Value {expected}")
```

```python
chi2_stat, p_value = chisquare(class_counts, expected)

print(f"Chichi2 Statistic: {chi2_stat}  \nP-value          : {p_value}")
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The distribution of passengers across classes is not uniform.")      
else:
    print("Fail to reject the null hypothesis: The distribution of passengers across classes is uniform.")
```

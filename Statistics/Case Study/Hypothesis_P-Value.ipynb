```python
import numpy as numpy
import pandas as pd
```

```python
train = pd.read_csv('Titanic Dataset/train.csv')
test  = pd.read_csv('Titanic Dataset/test.csv')
```

```python
df = pd.concat([train, test]).sample(1309)
```

```python
df.head()
```

```python
pop = df['Age'].dropna()
```

```python
pop
```

```python
sample_age = pop.sample(25).values
```

```python
sample_age
```

```python
# H0: the mean age is 35
# H1: the mean age is less than 35    
```

```python
from scipy.stats import shapiro
```

```python
shapiro_age = shapiro(sample_age)
shapiro_age
```

```python
sample_age.mean()
```

```python
pop_mean = 35
alpha = 0.05
```

```python
from scipy import stats
```

```python
t_stat, p_value = stats.ttest_1samp(sample_age, pop_mean)
print("T-statistic:", t_stat)
print("P-value:", p_value / 2)  # one-tailed test

if p_value / 2 < alpha and t_stat < 0:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
pop_male = df[df['Sex'] == 'male']['Age'].dropna()
pop_female = df[df['Sex'] == 'female']['Age'].dropna()
```

```python
sample_male = pop_male.sample(30).values
sample_female = pop_female.sample(30).values    

alpha = 0.05
```

```python
# H0: mean age of males = mean age of females
# H1: mean age of males != mean age of females
```

```python
from scipy.stats import shapiro
from scipy.stats import levene
```

```python
shapiro_male = shapiro(sample_male)
shapiro_female = shapiro(sample_female)

print("Shapiro-Wilk Test for Males  : W =", shapiro_male.statistic, ", p-value =", shapiro_male.pvalue)
print("Shapiro-Wilk Test for Females: W =", shapiro_female.statistic, ", p-value =", shapiro_female.pvalue)
```

```python
levene_test = levene(sample_male, sample_female)
print("Levene's Test: W =", levene_test.statistic, ", p-value =", levene_test.pvalue)
```

```python
t_stat, p_value = stats.ttest_ind(sample_male, sample_female)
print("T-statistic:", t_stat)
print("P-value    :", p_value)
```

```python
print("Mean Age of Males    :", sample_male.mean())
print("Mean Age of Females  :", sample_female.mean())
```

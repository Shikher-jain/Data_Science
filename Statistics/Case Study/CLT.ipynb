```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('Titanic Dataset/train.csv')
test_df = pd.read_csv('Titanic Dataset/test.csv')
```

```python
df = pd.concat([train_df.drop(columns=['Survived']), test_df]).sample(1309)
```

```python
df.head()
```

```python
df["Fare"].plot(kind = 'kde')
```

```python
samples = []
for i in range(1000):
    
    sample = df["Fare"].dropna().sample(50).to_list()
    samples.append(sample)

```

```python
samples = np.array(samples)
sample_means = samples.mean(axis=1)
sample_means
```

```python
sns.kdeplot(sample_means)
```

```python
sample_means.mean()
```

```python
lower_bound = sample_means.mean() - 1.96 * (sample_means.std() / np.sqrt(50))
upper_bound = sample_means.mean() + 1.96 * (sample_means.std() / np.sqrt(50))

print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
print(f"Population Mean: {df['Fare'].mean()}")
```

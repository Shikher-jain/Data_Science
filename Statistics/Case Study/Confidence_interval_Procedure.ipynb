```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('Titanic Dataset/train.csv')
test_df = pd.read_csv('Titanic Dataset/test.csv')
```

```python
df = pd.concat([train_df.drop(columns=['Survived']), test_df]).sample(1309)
```

```python
df.head()
```

```python
df["Fare"].plot(kind = 'kde')
```

```python
samples = []

std= []
for i in range(1000):
    
    sample = df["Fare"].dropna().sample(30).values
    std.append(sample.std())  
    samples.append(sample.tolist())
```

```python
samples = np.array(samples)

sample_means = samples.mean(axis=1)
sample_stds = np.mean(std)
sample_means, sample_stds
```

```python
samples.mean(axis=1), 
```

```python
sns.kdeplot(sample_means)
legend = ['Sample Means', 'Sample Standard Deviations']
plt.legend(legend)
```

```python
sample_means.mean()
```

```python
lower_bound = sample_means.mean() - 2.042 * (sample_stds/np.sqrt(30))
upper_bound = sample_means.mean() + 2.042 * (sample_stds/np.sqrt(30))

print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
print(f"Population Mean: {df['Fare'].mean()}")
```

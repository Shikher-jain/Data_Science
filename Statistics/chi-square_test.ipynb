```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sample1 = np.random.normal(loc=0, scale=1, size=1000)
sample2 = np.random.normal(loc=0, scale=1, size=1000)
sample3 = np.random.normal(loc=0, scale=1, size=1000)
sample4 = np.random.normal(loc=0, scale=1, size=1000)
sample5 = np.random.normal(loc=0, scale=1, size=1000)
```

```python
sample1,sample2
```

```python
x1 = sample1**2
x2 = sample2**2
x3 = sample3**2
x4 = sample4**2
x5 = sample5**2

y = x1 + x2
z = y + x3 
u = z + x4 
v = u + x5
```

```python
# sns.kdeplot(sample1, clip=(sample1.min(),sample1.max()))
# sns.kdeplot(sample2, clip=(sample2.min(),sample2.max()))
# sns.kdeplot(x2, clip=(x2.min(),x2.max()))

sns.kdeplot(x1, clip=(x1.min(),x1.max()), label='X1 = Sample1^2',legend=True)
sns.kdeplot(y, clip=(y.min(),y.max()), label='Y = X1 + X2',legend=True)
sns.kdeplot(z, clip=(z.min(),z.max()), label='Z = Y + X3 + X4 + X5',legend=True)
sns.kdeplot(u, clip=(u.min(),u.max()), label='U = Z + X4',legend=True)
sns.kdeplot(v, clip=(v.min(),v.max()), label='V = U + X5',legend=True)

legend = ['X1 = Sample1^2', 'Y = X1 + X2', 'Z = Y + X3 + X4 + X5', 'U = Z + X4', 'V = U + X5']
plt.legend(legend)
```

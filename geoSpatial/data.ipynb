```python
import pandas as pd
import numpy as np
```

```python
latlong = pd.read_csv(r"C:\shikher_jain\Data_Science\geoSpatial\district wise centroids.csv")
latlong.head()
```

```python
census = pd.read_csv(r"C:\shikher_jain\Data_Science\geoSpatial\india-districts-census-2011.csv")
census.head()
```

```python
df = latlong.merge(census, left_on='District', right_on='District name')
df.head()
```

```python
df = df.drop(columns=['District name', 'State name'])
df
```

```python
df['sex_ratio'] = round(df['Female'] / df['Male'] * 100)
df['literacy_rate'] = round(df['Literate'] / df['Population'] * 100)
df
```

```python
df.to_csv(r"C:\shikher_jain\Data_Science\geoSpatial\india.csv", index=False)
```

```python
import torch
import torch.nn as nn
```

### Conv2d

```python
covn = nn.Conv2d(in_channels=1, out_channels=8, kernel_size=3, stride=1, padding=1)
print(covn)
image = torch.randn(1, 1, 28, 28)
# print("Input  : ",image)
output = covn(image)
print("Input  : ",image.shape)
print("Output : ",output.shape)
```

```python
covn = nn.Conv2d(in_channels=1, out_channels=8, kernel_size=3, stride=2, padding=1)
print(covn)
image = torch.randn(1, 1, 28, 28)
# print("Input  : ",image)
output = covn(image)
print("Input  : ",image.shape)
print("Output : ",output.shape)
```

```python
covn = nn.Conv2d(in_channels=1, out_channels=8, kernel_size=3, stride=2, padding=0)
print(covn)
image = torch.randn(1, 1, 28, 28)
# print("Input  : ",image)
output = covn(image)
print("Input  : ",image.shape)
print("Output : ",output.shape)
```

```python
covn = nn.Conv2d(in_channels=1, out_channels=8, kernel_size=3, stride=1, padding=0)
print(covn)
image = torch.randn(1, 1, 28, 28)
# print("Input  : ",image)
output = covn(image)
print("Input  : ",image.shape)
print("Output : ",output.shape)
```

### MaxPool2d

```python
x = torch.randn(1, 8, 28, 28)
pool = nn.MaxPool2d(kernel_size=2, stride=2)
print(pool)
output = pool(x)
print("Input  : ",x.shape)
print("Output : ",output.shape)
```

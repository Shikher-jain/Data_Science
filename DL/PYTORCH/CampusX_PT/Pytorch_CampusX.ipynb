```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)

w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
w,y,w,b
```

```python
z = w*x+b
y_pred = torch.sigmoid(z)
loss = -(y*torch.log(y_pred) + (1-y)*torch.log(1-y_pred))
```

```python
z,y_pred,loss
```

```python
loss.backward()
```

```python
w.grad, b.grad
```

```python
import torch

x = torch.tensor(3.0, requires_grad=True)
w = torch.tensor(2.0)
y_pred = x * w

y_true = torch.tensor(10.0)
loss = (y_pred - y_true)**2
loss.backward()    
print(x.grad)       # tensor(4.)
```

```python
x = torch.tensor(2.0) # Input
w = torch.tensor(3.0, requires_grad=True) # Learnable Parameter

y_true = torch.tensor(10.0) # Target
lr =0.1 # learning rate

y_pred = x * w       # Predicted Output
y_true = torch.tensor(10.0)

loss = (y_pred - y_true)**2

loss.backward()    

print("Before Update")
print("w        :", w.item())  # tensor(3.0)
print("Gradient :", w.grad)       # tensor(-16.0)
print("loss     :", loss.item())

with torch.no_grad():
    w -= lr * w.grad  # Update Rule
w.grad.zero_()    # Zero the Gradients

print("After Update")
print("w        :", w.item())  # tensor(3.0)
print("Gradient :", w.grad)       # tensor(-16.0)
print("loss     :", loss.item())
```

```python
x = torch.tensor([1., 2., 3., 4., 5., 6.]) # Input
y_true = torch.tensor([2., 4., 6., 8., 10., 12.]) # Target

w = torch.tensor(0.0, requires_grad=True)  # Initialize weight
lr = 0.01  # Learning rate

epochs =30
for epoch in range(epochs):
    y_pred = x * w                       # Forward Pass
    loss = ((y_pred - y_true) ** 2).mean()  # Mean Squared Error Loss

    loss.backward()                      # Backward Pass

    with torch.no_grad():
        w -= lr * w.grad                 # Update weight
    w.grad.zero_()                       # Zero the gradients

    print(f'Epoch {epoch+1}: w = {w.item():.4f}, loss = {loss.item():.4f}')
```

### Implementing Tensors in PyTorch

y = w * x + b

```python
x = torch.tensor([1.,2.,3.,4.,5.])
y = torch.tensor([2.,  4.,  6.,  8., 10.])
```

```python
w = torch.tensor(0.0)
b = torch.tensor(0.0)
```

```python
y_pred = w * x + b
```

```python
loss = ((y_pred - y)**2).mean()
```

```python
print("Actual Value",y)
print("Predicted Value",y_pred)
print("Loss:",loss)
print("Loss:",loss.item())
```
